# Extraer nueva informacion de la descripcion

**TextBlob aims to provide access to common text-processing operations through a familiar interface. You can treat [TextBlob](https://textblob.readthedocs.io/en/dev/quickstart.html) objects as if they were Python strings that learned how to do Natural Language Processing.**

```python
# pip install -U textblob
# python -m textblob.download_corpora
from textblob import TextBlob

text = '''
The titular threat of The Blob has always struck me as the ultimate movie
monster: an insatiably hungry, amoeba-like mass able to penetrate
virtually any safeguard, capable of--as a doomed doctor chillingly
describes it--"assimilating flesh on contact.
Snide comparisons to gelatin be damned, it's a concept with the most
devastating of potential consequences, not unlike the grey goo scenario
proposed by technological theorists fearful of
artificial intelligence run rampant.
'''

blob = TextBlob(text)

print(blob.tags[:4], '...')

print(blob.noun_phrases[:4], '...')

for sentence in blob.sentences:
    print(sentence.sentiment.polarity)
```

**Decodificar los html entities de las descripciones**

```python
import html

desc = df[df['metrostotales'].isnull()]['descripcion'].iloc[1]

print('antes del unescape:\n', desc, '\n')
desc = html.unescape(desc) #esto habria que hacerlo con todas las descripciones y borrar los <p>, etc
print('despues:\n', desc, '\n')
```

**Paso el ejemplo por textBlob**

The sentiment property returns a namedtuple of the form Sentiment(polarity, subjectivity). The polarity score is a float within the range [-1.0, 1.0]. The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.
El problema es que al parecer no esta entrenado para español https://github.com/sloria/TextBlob/issues/209

[Tag list](https://www.geeksforgeeks.org/python-part-of-speech-tagging-using-textblob/)

```python
blob = TextBlob(desc)

print(blob.tags[:8], '...\n')

print(blob.noun_phrases, '\n')

for sentence in blob.sentences:
    print(sentence, sentence.sentiment.polarity)
```

```python
from opencage.geocoder import OpenCageGeocode

key = '24e5357cc1ef4b02bd3dbe9d828f83ba'
key = 'b431379f19dc4affaae52668e004fbae'
geocoder = OpenCageGeocode(key)

def getLatLongOpencage(row):
    results = geocoder.geocode(f"{row.direccion}, {row.ciudad}, {row.provincia}, Mexico")
    lat = lng = None
    
    if results and len(results):
        lat = results[0]['geometry']['lat']
        lng = results[0]['geometry']['lng']
    else:
        print(f'ERROR para id {row.id}')
        
    return pd.Series([lat,lng])
```

```python
import requests
# http://api.positionstack.com/v1/forward
#     ? access_key = YOUR_ACCESS_KEY
#     & query = 1600 Pennsylvania Ave NW, Washington DC

latMin = round(df.lat.min())
latMax = round(df.lat.max())
lngMin = round(df.lng.min())
lngMax = round(df.lng.max())

def getLatLongPositionStack(row):
    address = f"{row.direccion}, {row.ciudad}, {row.provincia}, Mexico"
    params = {'access_key':'00d72a5f857ac8c8d152393b465413b3', 'query': address}
    response = requests.get('http://api.positionstack.com/v1/forward', params=params)
    response
    
    lat = lng = None
    
    if response.ok:
        for place in response.json()['data']:       
            if len(place) == 0:
                continue
                
            placeLat = response.json()['data'][0]['latitude']
            placeLng = response.json()['data'][0]['longitude']
            if row.ciudad in place['label'] or (place['locality']!=None and row.ciudad in place['locality']): #prefiero que sea de la misma ciudad
                lat = placeLat
                lng = placeLng
                
            elif lat == None and (row.provincia in place['label'] or (place['region']!=None and row.provincia in place['region'])): #sino me conformo con menos
                lat = placeLat
                lng = placeLng
                
            elif lat == None and ('Mexico' in place['label'] or (place['country']!=None and 'Mexico' in place['country'])): #quedo horrible pero anda
                lat = placeLat
                lng = placeLng
                
            elif latMin <= placeLat <= latMax and lngMin <= placeLng <= lngMax:
                lat = placeLat
                lng = placeLng
    else:
        print(f'ERROR response invalida con {row.id}')
        
    if not lat:
        print(f'ERROR no se pudo encontrar lat,lng para {row.id}')
        
    return pd.Series([lat,lng])
```

```python
batchSize = 50
desde = 5000
limite = desde + 5000

while desde < limite:
    sinLat.iloc[desde:desde+batchSize][['lat', 'lng']] = sinLat.iloc[desde:desde+batchSize].apply(lambda row: getLatLongPositionStack(row), axis=1)
    desde += batchSize
    print(desde) # quiero ver por donde vamos

    latLngRecuDF.update(sinLat)
    latLngRecuDF.to_csv('lat_lng_recuperados1.csv', index=False) # voy haciendo de a tandas para no perder datos en caso que falle
    
print('FIN')
```
